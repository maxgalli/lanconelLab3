```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import gridspec            # to arrange the plots nicely
import pandas as pd
import math
from scipy.optimize import curve_fit

%matplotlib inline
plt.style.use('tableau-colorblind10')
```

```python
# preliminar

index = []
for count in range(0,15):
    index.append(count)
    
sizes = [8.0,6.3,5.0,4.0,3.2,2.5,2.0,1.6,1.3,1.0,0.8,0.6,0.5,0.4,0.3]

df_dict_in = {}
df_dict_out = {}
```

```python
# 2.5

for count in range(0,15):
    df_dict_in['df%i'%(count)] = pd.read_csv('25_in_%i.csv'%(count), sep=' ', names=['stuff1','stuff2','mean_in','std_in'])
    df_dict_out['df%i'%(count)] = pd.read_csv('25_out_%i.csv'%(count), sep=' ', names=['stuff1','stuff2','mean_out','std_out'])

df_in = df_dict_in['df0']
df_out = df_dict_out['df0']
for count in range(1,15):
    df_in = pd.concat([df_in,df_dict_in['df%i'%(count)]])
    df_out = pd.concat([df_out,df_dict_out['df%i'%(count)]])
    
df_in = df_in[['mean_in','std_in']]
df_out = df_out[['mean_out','std_out']]
df_in.index = index
df_out.index = index

df = pd.concat([df_in,df_out], axis=1)

df['cnr']=(df['mean_in']-df['mean_out'])/df['std_out']

df['sizes']=sizes

# plot
plt.ylabel('CNR')
plt.xlabel('Thickness [mm]')
plt.plot(df.sizes, df.cnr, 'go-')
plt.savefig('25_plot.png')

# for other plot
mean25 = df['cnr'].mean()
std25 = df['cnr'].std()

print(mean25)
print(std25)
```
